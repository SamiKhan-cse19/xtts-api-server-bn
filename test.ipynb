```python
import torch
import torchaudio
from TTS.tts.configs.xtts_config import XttsConfig
from TTS.tts.models.xtts import Xtts
from banglanlptoolkit import BnNLPNormalizer

normalizer = BnNLPNormalizer()

config = XttsConfig()
config.load_json("xtts_models/v2.0.2/config.json")
model = Xtts.init_from_config(config)
model.load_checkpoint(config, checkpoint_dir="xtts_models/v2.0.2/")
model.cuda()
```

```python
reference_audio = "example/female.wav"
text = "আমি বাংলায় গান গাই।"
out = model.synthesize(
    text,
    config,
    speaker_wav=reference_audio,
    gpt_cond_len=3,
    language="bn",
    speed=1.0,
)
```

```python
from IPython.display import Audio

torchaudio.save("xtts.wav", torch.tensor(out["wav"]).unsqueeze(0), 24000)
Audio("xtts.wav")
```

```python
gpt_cond_latent, speaker_embedding = model.get_conditioning_latents(audio_path=reference_audio)

chunks = model.inference_stream(
    text,
    speaker_wav=reference_audio,
    gpt_cond_len=3,
    gpt_cond_latent=gpt_cond_latent,
    speaker_embedding=speaker_embedding,
    language="bn",
    speed=1.0,
)
```

```python
for chunk in chunks:
    print(chunk)
    torchaudio.save("xtts_stream.wav", torch.tensor(chunk["wav"]).unsqueeze(0), 24000)
    Audio("xtts_stream.wav")
```
